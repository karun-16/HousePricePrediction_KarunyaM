# Task 3: Model Building

The objective of this task is to build machine learning models capable of predicting house prices using the cleaned housing dataset. The models will be trained and evaluated to compare their predictive performance.

### Requirement 1: Split the Data into Training and Test Sets

The cleaned dataset is divided into training and testing subsets. The training set is used to train the machine learning models, while the testing set is reserved for evaluating model performance on unseen data.

```python
from sklearn.model_selection import train_test_split
```

```python
# Separating features and target variable

X = df.drop("price", axis=1)
y = df["price"]
```

```python
# Splitting the dataset

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

print("Training and testing datasets created successfully.")
```

```python
print("Training set:", X_train.shape)
print("Testing set:", X_test.shape)
```

### Observation

The cleaned dataset was successfully divided into training and testing sets using an 80:20 ratio. The training data contains 436 observations, while the testing data contains 109 observations for evaluating model performance.

### Requirement 2: Train a Linear Regression Model

A Linear Regression model is trained using the training dataset to learn the relationship between the housing features and the selling price. The trained model is then used to predict house prices.

```python
from sklearn.linear_model import LinearRegression
```

```python
# Creating the Linear Regression model

lr_model = LinearRegression()

# Training the model

lr_model.fit(X_train, y_train)

print("Linear Regression model trained successfully.")
```

```python
# Predicting house prices

y_pred_lr = lr_model.predict(X_test)

print("House price predictions generated successfully.")
```

```python
# Displaying sample predictions

print("Sample predictions:")
print(y_pred_lr[:5])
```

### Observation

The Linear Regression model was successfully trained using the training dataset. The model generated house price predictions for the testing dataset, indicating that it has learned the relationship between the housing features and the target variable.

### Requirement 3: Evaluate the Linear Regression Model

The performance of the Linear Regression model is evaluated using standard regression metrics to measure prediction accuracy and overall model effectiveness.

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# Calculating evaluation metrics
import numpy as np

mae_lr = mean_absolute_error(y_test, y_pred_lr)
rmse_lr = np.sqrt(mean_squared_error(y_test, y_pred_lr))
r2_lr = r2_score(y_test, y_pred_lr)
```

```python
# Displaying evaluation metrics

print("Mean Absolute Error (MAE):", mae_lr)
print("Root Mean Squared Error (RMSE):", rmse_lr)
print("R² Score:", r2_lr)
```

### Observation

The Linear Regression model was evaluated using MAE, RMSE, and R² score. The model achieved an MAE of approximately 970,043, an RMSE of approximately 1,324,507, and an R² score of 0.653. The R² value indicates that the model explains about 65% of the variation in house prices, providing a reasonable baseline for comparison with other machine learning models.

## Requirement 4: Train a Random Forest Regressor

A Random Forest Regressor is trained using the training dataset to capture complex relationships between housing features and house prices. The trained model is then used to generate predictions for the testing dataset.


```python
from sklearn.ensemble import RandomForestRegressor
```

```python
# Creating the Random Forest model

rf_model = RandomForestRegressor(
    n_estimators=100,
    random_state=42
)

# Training the model

rf_model.fit(X_train, y_train)

print("Random Forest Regressor trained successfully.")
```

```python
# Predicting house prices

y_pred_rf = rf_model.predict(X_test)

print("House price predictions generated successfully.")
```

```python
# Displaying sample predictions

print(y_pred_rf[:5])
```

### Observation

The Random Forest Regressor was successfully trained using the training dataset and generated predictions for the testing dataset. The model can capture complex patterns in the housing data and will be compared with the Linear Regression model to determine which provides better prediction performance.

## Requirement 5: Evaluate the Random Forest Regressor

The performance of the Random Forest Regressor is evaluated using MAE, RMSE, and R² score to measure prediction accuracy and compare it with the Linear Regression model.

```python
# Calculating evaluation metrics

mae_rf = mean_absolute_error(y_test, y_pred_rf)
rmse_rf = mean_squared_error(y_test, y_pred_rf) ** 0.5
r2_rf = r2_score(y_test, y_pred_rf)
```

```python
# Displaying evaluation metrics

print("Mean Absolute Error (MAE):", mae_rf)
print("Root Mean Squared Error (RMSE):", rmse_rf)
print("R² Score:", r2_rf)
```

### Observation

The Random Forest Regressor was evaluated using MAE, RMSE, and R² score. The model achieved an MAE of approximately 1,021,546, an RMSE of approximately 1,400,566, and an R² score of 0.612. These results will be compared with those of the Linear Regression model to determine the better-performing approach for this dataset.

## Requirement 6: Compare Model Performance

The performance of the Linear Regression and Random Forest Regressor models is compared using MAE, RMSE, and R² score to identify the better-performing model for house price prediction.

```python
# Creating a comparison table

comparison = pd.DataFrame({
    "Metric": ["MAE", "RMSE", "R² Score"],
    "Linear Regression": [mae_lr, rmse_lr, r2_lr],
    "Random Forest": [mae_rf, rmse_rf, r2_rf]
})

comparison
```

### Observation

The performance comparison shows that the Linear Regression model achieved lower MAE and RMSE values and a higher R² score than the Random Forest Regressor. Based on these evaluation metrics, Linear Regression provided better prediction performance for this housing dataset and is the preferred model among the two approaches tested.
